```python
df10_cleaned = df1000[df1000['Rank']<300000]
```

```python
pplot10_cleaned = df10_cleaned[['Price','Fees','Net','Rank','Sellers','LQS','Sales','Revenue','Rated','X of Rating']]
pplot10 = pplot10_cleaned.dropna() #dropping nan values
```

```python
sns.set(style='darkgrid')
g = sns.pairplot(pplot10,kind='reg',plot_kws ={'scatter_kws': {'alpha': 0.5}})

plt.title('Pairplot of Sales under 1000 and Rankings below 300,000')
plt.show()
```

```python
import numpy as np
sns.set(style='white')
#finding a correlation matrix
corr = pplot10.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
# Set up the matplotlib figure

f, ax = plt.subplots(figsize=(18,10))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
_ = sns.heatmap(corr, mask = mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.title('Correlation Matrix of Features')
plt.show()
```

```python
#importing PCA to perform dimension reduction on dataset
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

scaler = StandardScaler()  #instantiate a scaler object
pca = PCA()              #instantiate a PCA object
pipeline = make_pipeline(scaler,pca) #create a pipeline that feeds into scaler then PCA
```

```python
pipeline.fit(pplot10) #fitting PCA model on data
```

```python
#create a plot that shows PCA features and variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.title('Feature Variance')
plt.show()
```

```python
#creating a pca with 7 dimensions instead of 8 
pca = PCA(n_components=7)
pca.fit(pplot10)
pca_features=pca.transform(pplot10)
```
